```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```
Data = pd.read_csv('dataset.csv')
Data.head(5)
```

```
Data.info()
```

```
Top10 = Data.sort_values('popularity', ascending=False)
Top10.head(10)
```

```
dop = Top10.drop_duplicates(subset=['track_name'])

dop.head(10)
```

```


popularidad_artista = dop.groupby('artists')['popularity'].sum().reset_index()

top10 = popularidad_artista.sort_values(by='popularity', ascending=False)

top10 = top10.head(10)

plt.figure(figsize=(12, 6))
plt.barh(top10['artists'], top10['popularity'], color='skyblue')
plt.xlabel('Popularidad')
plt.ylabel('Artistas')
plt.title('Los 10 artistas más populares')
plt.gca().invert_yaxis()
plt.show()
```

```
popularidad_artista = dop.groupby('track_genre')['popularity'].sum().reset_index()

top10 = popularidad_artista.sort_values(by='popularity', ascending=False)

top10 = top10.head(10)

plt.figure(figsize=(12, 6))
plt.barh(top10['track_genre'], top10['popularity'], color='skyblue')
plt.xlabel('Popularidad')
plt.ylabel('Artistas')
plt.title('Los 10 artistas más populares')
plt.gca().invert_yaxis()
plt.show()
```
